```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\hp\Downloads\1000 Sales Records.csv' , encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
#CHECK NULL VALUES
pd.isnull(df).sum()
```

```python
#CHANGE DATA TYPE
df['Total Cost'] = df['Total Cost'].astype('int')
```

```python
df['Total Cost'].dtype
```

```python
df.describe()
```

```python
ax = sns.countplot(y = 'Region', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.countplot(data = df, y = 'Item Type')
```

```python
sales_country = df.groupby(['Country'], as_index = False)['Total Profit'].sum().sort_values(by= 'Total Profit', ascending= False).head(12)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_country, x = 'Country', y='Total Profit')
```

```python
ax = sns.countplot(x = 'Sales Channel', data = df)
sns.set(rc={'figure.figsize':(5,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_country = df.groupby(['Item Type'], as_index = False)['Total Revenue'].sum().sort_values(by= 'Total Revenue', ascending= False).head(12)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_country, x = 'Item Type', y='Total Revenue')
```

```python
Unit_Sold = df.groupby(['Units Sold'], as_index = False)['Region'].sum().sort_values(by= 'Region', ascending= False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = Unit_Sold, x = 'Units Sold', y='Region')
```

```python
sales_country = df.groupby(['Item Type'], as_index = False)['Total Profit'].sum().sort_values(by= 'Total Profit', ascending= False).head(12)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_country, x = 'Item Type', y='Total Profit')
```
